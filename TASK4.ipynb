TASK 4 : EMAIL SPAM DETECTION

```python
import pandas as pd
data = pd.read_csv("spam.csv",encoding='ISO-8859-1')
data.head()
```

```python
data.info()
data.describe()
```

```python
data['category'].value_counts()
```

KNeighborsClassifier is performed to classify spam detection

```python
print(data)
```

```python
# replace the null values with a null string
mail_data = data.where((pd.notnull(data)),'')
```

```python
mail_data.head()
```

spam as 0 and ham as 1

```python
mail_data.loc[mail_data['category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['category'] == 'ham', 'Category',] = 1
```

```python
# separate text and label
x = mail_data['messages']
y = mail_data['category']
```

```python
print(x)
```

```python

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import KNeighborsClassifier


xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
print(x.shape)
print(xtrain.shape)
print(xtest.shape)
```

```python
# Create a CountVectorizer to convert text to numerical features
cv = CountVectorizer()
feature_train = cv.fit_transform(xtrain)
feature_test = cv.transform(xtest)

# Initialize and train the KNN classifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(feature_train,ytrain)

# Evaluate the KNN model
accuracy = model.score(feature_test, ytest)
print("Accuracy:", accuracy)
```

Hence the model accuracy is 91%

```python
mail = ["Go until jurong point, crazy.. Available only in bugis n great world la e buffet... Cine there got amore wat..."]

input_features =cv.transform(mail)

# making prediction
prediction = model.predict(input_features)
if (prediction[0]==1):
  print('Ham mail')
else:
  print('Spam mail')
```

hence it is correctly detecting spam or not
